#### **Data Description**

The dataset focuses on movies and TV shows available on Amazon Prime Video and is created by combining two separate data sources: `titles.csv` and `credits.csv`.

The `titles.csv `file provides detailed information about each movie or TV show on the platform. Every row in this dataset represents a single title along with its related attributes such as release year, genre, runtime, and ratings.

The `credits.csv` file contains information about the people associated with these titles, including actors and directors. Each row represents an individual’s role and involvement in a specific movie or TV show.

#### Missing Values and Null Values

```python
# null values
df.isnull().sum()
```

There are null values in description column, So drop all the null values from description column.

```python
# dropping null values from description
df.dropna(subset = ['description'], inplace = True)
```

```python
#Mode of age_certification
df['age_certification'].mode()[0]
```

```python
#Replace all null values of age_certification with mode
df['age_certification'].fillna(df['age_certification'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

The `seasons` column contains missing values because this information is relevant only for TV shows. Movies do not have seasons, so these entries are expected to be empty. Instead of removing such records, the missing values are replaced with 0 to correctly indicate that movies have no seasons.

```python
#Replace null values in seasons column with 0
df['seasons'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

Here `imdb_id `serves as a unique identifier for both movies and TV shows, records with missing IMDb IDs are not useful for reliable analysis. Therefore, all rows containing null values in the `imdb_id` column are removed.

```python
df.dropna(subset = ['imdb_id'], inplace = True)
```

```python
df.isnull().sum()
```

The `imdb_score` column represents the rating of a movie or TV show. Removing rows with missing values in this column could result in unnecessary data loss, so the missing values are instead replaced with the mean IMDb score to preserve the overall dataset.

```python
round(df['imdb_score'].mean(),1)
```

```python
#Replace null values in imdb_score with mean of imdb_score
df['imdb_score'].fillna(round(df['imdb_score'].mean(),1), inplace = True)
```

```python
df.shape
```

The `imdb_votes` column indicates how many votes a movie or TV show has received. Missing values in this column may suggest that a title is either newly released or has not yet attracted significant audience attention. Therefore, these missing values are replaced with 0 to reflect the absence of votes rather than removing the records.

```python
df['imdb_votes'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
#Mean of tmdb_popularity
round(df['tmdb_popularity'].mean(),2)
```

```python
# replace null values in 'tmdb_popularity' column with mean
df['tmdb_popularity'].fillna(round(df['tmdb_popularity'].mean(),2), inplace =True)
```

```python
#Mean of tmdb_score
round(df['tmdb_score'].mean(),1)
```

```python
# replace null values in 'tmdb_score' column with mean
df['tmdb_score'].fillna(round(df['tmdb_score'].mean(),1), inplace = True)
```

Missing values in the `character` column indicate that details about the role played by a person in a movie or TV show are not available in the dataset. To handle this, these missing entries are replaced with `'unknown'`.

```python
df['character'].fillna('unknown', inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

### 1. **Histogram**

Histograms are used to visualize the distribution of numerical variables. They provide a clear view of how data values are spread across different ranges, helping to understand the shape of the data distribution, such as whether it is skewed, normal, or contains outliers.

#### Histogram of release_year

```python
#Histogram of release_year
plt.figure(figsize = (8,6))
sns.histplot(data = df, x = 'release_year', bins = 20, color = 'red', kde = True)
plt.title('Histogram of release year\n', color = 'black')
plt.show()
```

**Insights**

* The histogram of the `release_year` column illustrates how movies and TV shows are distributed across different years. A significant portion of the content available on Amazon Prime Video has been released in the last decade, indicating a strong focus on recent productions.

* The distribution is left-skewed, showing fewer titles from earlier years and a steady increase in releases over time. This overall upward trend highlights the continuous expansion of Amazon Prime Video’s content library as the platform adds more titles each year.

#### Histogram of runtime

```python
# Histogram of runtime
plt.figure(figsize = (8,6))
sns.histplot(data = df, x = 'runtime', bins = 20, color = 'green', kde = True)
plt.title('Histogram of runtime\n', color = 'black')
plt.show()
```

**Insight**

* The histogram shows that the runtime of movies and TV shows on Amazon Prime Video does not follow a normal distribution. Instead, it is right-skewed, with a longer tail extending toward higher runtimes.

* The highest concentration of titles falls within the 70–90 minute range, indicating that this is the most common runtime on the platform. Additionally, a large number of movies and shows have runtimes between 60 and 100 minutes, suggesting this duration may be ideal for maintaining viewer interest while allowing effective storytelling.

#### Histogram of seasons

```python
#Histogram of Seasons
plt.figure(figsize = (8,6))
sns.histplot(data = df, x = 'seasons', bins = 20, color = 'yellow')
plt.title('Histogram of seasons\n', color = 'black')
plt.show()
```

**Insights**

#### Histogram of imdb score

```python
#Histogram of imdb_score
plt.figure(figsize = (8,6))
sns.histplot(data = df, x = 'imdb_score', bins = 30, color = 'orange', kde = True)
plt.title('Histogram of imdb_score\n', color = 'black')
plt.show()
```

### **Bivariate Analysis**

Bivariate analysis focuses on examining the relationship between two variables to understand how one variable may influence or relate to another.

**1. Scatter Plot**

A scatter plot is a visualization technique that represents the relationship between two numerical variables using individual data points. Each dot corresponds to a pair of values, making scatter plots useful for identifying patterns, trends, correlations, or outliers between variables.

Scatter plot of imdb_vote and imdb_scores

```python
#Scatter plot of imdb_vote and imdb_scores
sns.scatterplot(data = df, x = 'imdb_votes', y = 'imdb_score', color = 'green')
plt.title('Scatter plot of imdb_vote and imdb_scores\n', color = 'black')
plt.show()
```

**Insights**

* The scatter plot shows a positive relationship between IMDb votes and IMDb scores, indicating that titles with a higher number of votes generally tend to have higher ratings.

* This positive correlation suggests that more popular titles—those receiving greater audience attention—are often rated more favorably, reflecting stronger viewer engagement and satisfaction.

**2. Line Plot**

A line plot is used to connect data points with a continuous line, making it effective for visualizing trends and patterns over time. It is especially useful for observing how a numerical variable changes with respect to another continuous variable, such as time.

```python
#line plot of runtime v/s release_year
plt.figure(figsize = (8,6))
sns.lineplot(data = df, y = 'runtime', x = 'release_year', color = 'green')
plt.title('line plot of runtime v/s release_year\n', color = 'black')
plt.show()
```

**Insight**

* The line plot shows a slight downward trend in the average runtime of movies and TV shows over the years, indicating that content on Amazon Prime Video has gradually become shorter.

* Earlier releases generally had longer runtimes, whereas more recent titles tend to be more concise. This trend may be influenced by factors such as shorter viewer attention spans, changing content consumption habits, and production cost considerations.

```python
#line plot of seasons v/s release_year
plt.figure(figsize = (8,6))
sns.lineplot(data = df, y = 'seasons', x = 'release_year', color = 'red')
plt.title('line plot of seasons v/s release_year\n', color = 'black')
plt.show()
```

**Insights**

* The line plot shows an upward trend in the number of seasons for TV shows over the years, indicating that series on Amazon Prime Video are increasingly extending across multiple seasons.

* This trend suggests that viewers are willing to invest more time in long-running shows, reflecting sustained interest and engagement with episodic content.

```python
#line plot of imdb_score v/s release_year
plt.figure(figsize = (8,6))
sns.lineplot(data = df, y = 'imdb_score', x = 'release_year')
plt.title('line plot of imdb_score v/s release_year\n', color = 'black')
plt.show()
```

#### **4. Violin Plot**

A violin plot is a statistical visualization that combines features of a box plot and a kernel density plot. It is used to display the distribution of a numerical variable across different categories, providing insight into the data’s density, spread, and central tendency

```python
#Violin Plot of imdb_score by type
plt.figure(figsize=(10, 6))
sns.violinplot(x='type', y='imdb_score', data=df)
plt.title('Violin Plot of imdb_score by type')
plt.show()
```

**Insights**

* For Movies, the violin plot shows a higher concentration of IMDb scores between 6 and 7, indicating that most movies on Amazon Prime Video receive reasonably good ratings.

* For TV Shows, the score distribution is more concentrated between 7 and 8, suggesting that TV series on the platform generally achieve higher ratings compared to movies.

* The median IMDb score for movies is slightly lower than that of TV shows, as indicated by the central marker in the violin plot. This implies that, on average, TV shows tend to be better rated than movies on Amazon Prime Video
